```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corona_dataset_csv = pd.read_csv(r'C:\Users\Admin\Downloads\New folder\doc\Covid 19\Covid19_Confirmed_dataset.csv')
corona_dataset_csv.head(10)
```

```python
corona_dataset_csv.drop(["Lat","Long"], axis=1, inplace = True)
```

```python
corona_dataset_csv.head(10)
```

```python
corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum()
```

```python
#First derivative of spread
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
#find max infection rate for china
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
#find max infection rate for India
corona_dataset_aggregated.loc["India"].diff().max()
```

```python
#find max infection rate for US
corona_dataset_aggregated.loc["US"].diff().max()
```

```python
#Find max infection rate for all countries
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries :
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
max_infection_rates
```

```python
corona_dataset_aggregated["max_infection_rates"] = max_infection_rates
```

```python
corona_dataset_aggregated.head()
```

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rates"])
```

```python
corona_data.head()
```
